#split the data

```python
X = data.drop('Global_Sales',axis=1)
y = data['Global_Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=25)
```

#Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
linear_predict = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error,mean_absolute_percentage_error
import numpy as np
```

```python
#Metrics in Linear Regression to evaluate the performance
linear_mse = mean_squared_error(y_test,linear_predict)
linear_r2 = r2_score(y_test,linear_predict)
linear_rmse = np.sqrt(linear_mse)
linear_mae = mean_absolute_error(y_test,linear_predict)
linear_mape = mean_absolute_percentage_error(y_test,linear_predict)
# linear_mape = np.mean(np.abs((y_test-linear_predict)/y_test))*100
```

#Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)

x_train_poly = poly.fit_transform(X_train)
poly_model = LinearRegression()
poly_model.fit(x_train_poly,y_train)

x_test_poly = poly.fit_transform(X_test)
poly_predict = poly_model.predict(x_test_poly)
```

```python
#Metrics in Polynomial Regression to evaluate the performance
poly_mse = mean_squared_error(y_test,poly_predict)
poly_r2 = r2_score(y_test,poly_predict)
poly_rmse = np.sqrt(poly_mse)
poly_mae = mean_absolute_error(y_test,poly_predict)
poly_mape = mean_absolute_percentage_error(y_test,poly_predict)
```

#Lasso Regression(L1)

```python
from sklearn.linear_model import Lasso
lasso_model = Lasso()
```

```python
lasso_model.fit(X_train,y_train)
```

```python
lasso_predict = lasso_model.predict(X_test)
```

```python
#Metrics in Lasso Regression to evaluate the performance
lasso_mse = mean_squared_error(y_test,lasso_predict)
lasso_r2 = r2_score(y_test,lasso_predict)
lasso_rmse = np.sqrt(lasso_mse)
lasso_mae = mean_absolute_error(y_test,lasso_predict)
lasso_mape = mean_absolute_percentage_error(y_test,lasso_predict)
```

#Ridge Regression(L2)

```python
from sklearn.linear_model import Ridge
ridge_model = Ridge()
```

```python
ridge_model.fit(X_train,y_train)
```

```python
ridge_predict = ridge_model.predict(X_test)
```

```python
#Metrics in Ridge Regression to evaluate the performance
ridge_mse = mean_squared_error(y_test,ridge_predict)
ridge_r2 = r2_score(y_test,ridge_predict)
ridge_rmse = np.sqrt(ridge_mse)
ridge_mae = mean_absolute_error(y_test,ridge_predict)
ridge_mape = mean_absolute_percentage_error(y_test,ridge_predict)
```

```python
print('Linear Regression Metrics')
print('--------------------------')
print('MSE : ', linear_mse)
print('R2 : ', linear_r2)
print('RMSE : ', linear_rmse)
print('MAE : ', linear_mae)
print('MAPE : ', linear_mape)

print('\nPolynomial Regression Metrics')
print('--------------------------')
print('MSE : ', poly_mse)
print('R2 : ', poly_r2)
print('RMSE : ', poly_rmse)
print('MAE : ', poly_mae)
print('MAPE : ', poly_mape)

print('\nLasso Regression Metrics')
print('--------------------------')
print('MSE : ', lasso_mse)
print('R2 : ', lasso_r2)
print('RMSE : ', lasso_rmse)
print('MAE : ', lasso_mae)
print('MAPE : ', lasso_mape)

print('\nRidge Regression Metrics')
print('--------------------------')
print('MSE : ', ridge_mse)
print('R2 : ', ridge_r2)
print('RMSE : ', ridge_rmse)
print('MAE : ', ridge_mae)
print('MAPE : ', ridge_mape)
```

```python
linear_metrics = [linear_mse, linear_r2, linear_rmse, linear_mae, linear_mape]
poly_metrics = [poly_mse, poly_r2, poly_rmse, poly_mae, poly_mape]
lasso_metrics = [lasso_mse, lasso_r2, lasso_rmse, lasso_mae, lasso_mape]
ridge_metrics = [ridge_mse, ridge_r2, ridge_rmse, ridge_mae, ridge_mape]

index_metrics = ['MSE', 'R2', 'RMSE', 'MAE', 'MAPE']

data = {'Linear' : linear_metrics,
        'Polynomial' : poly_metrics,
        'Lasso' : lasso_metrics,
        'Ridge' : ridge_metrics}

df_metrics = pd.DataFrame(data, index=index_metrics)

df_metrics.index.name = 'Evaluation Metrics'
df_metrics.columns.name = 'Regression ML Models'

df_metrics
```
